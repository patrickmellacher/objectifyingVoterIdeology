# Replacing existing categories
Procedure (for alternative):
- For the three target columns lrgen, lrecon and galtan...
    - 0, 1, 2 (0-3) are replaced by 'l'
    - 3-7 (4-6) are replaced by 'n'
    - 8-10 (7-10) are replaced by 'r'

```python
# remove rows with excessive number of NAN's - at least 4 values out of 6 must be provided
if alternative_option == '434':
    data_base.iloc[:, -3:] = data_base.iloc[:, -3:].replace({0: 'l', 1: 'l', 2: 'l', 3: 'l', 
                                                             4: 'c', 5: 'c', 6: 'c',
                                                             7: 'r', 8: 'r', 9: 'r', 10: 'r'})
else:
    data_base.iloc[:, -3:] = data_base.iloc[:, -3:].replace({0: 'l', 1: 'l', 2: 'l', 
                                                             3: 'c', 4: 'c', 5: 'c', 6: 'c', 7: 'r', 
                                                             8: 'r', 9: 'r', 10: 'r'})
```
